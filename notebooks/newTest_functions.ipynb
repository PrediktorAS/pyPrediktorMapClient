```python
def get_object_id(sitename: str) -> str:
    """Function to get object id of an object from display name

    Args:
        display_name (str): Display name of a site/subsite
    """
    try:
        obj_type =  next(item for item in mdx.get_objects_of_type('SiteType') if item["DisplayName"] == sitename)
    except StopIteration:
        obj_type = {}
    object_id = [obj_type.get("Id")]
    return object_id
```

```python
def get_object_descendants(type_name: str, sitename: str, domain: str):
        """A function to get object descendants

        Args:
            type_name (str): type_name of a descendant
            object_Ids (List[str]): object id(s) of parent
            domain (str): PV_Assets or PV_Serves

        Returns:
            json data: descendats data of selected object id(s)
        """
        object_type_id = mdx.get_object_typeid(type_name)
        object_Ids = get_object_id(sitename)
        body = json.dumps({
            "typeId": object_type_id,
            "objectIds": object_Ids,
            "domain": domain
            })
        return mdx.request('POST', 'query/object-descendants', body)
```
